| **Complication** | Growth bought with traffic; unclear which levers drive *profitable* growth, and an apparent landing-page winner tempts a rollout. |

**Key Lines (MECE):** (1) Mobile is the biggest conversion lever · (2) Cross-sell of higher-margin newer products grows profit · (3) The landing-page winner is a confounding artifact — don't act on it.
