## Assemble the report body from the evidence

```python
def pct(x): return f"{x*100:.1f}%"
best_feat = K["best_single_feature"]; best_auc = K["best_single_auc"]
w = K["winner"]
tiles = [
    (f'{K["winner_cv_auc"]:.3f}', "CV ROC-AUC"),
    (pct(K["op_sensitivity"]), "malignant sensitivity"),
    (f'{cm["fn"]}/{cm["tp"]+cm["fn"]}', "malignancies missed (test)"),
    (f'{K["vif_over10"]}/30', "features collinear (VIF>10)"),
]
tiles_html = "".join(f'<div class="tile"><div class="n">{n}</div><div class="l">{l}</div></div>' for n,l in tiles)

# model table
import csv, io
rows = list(csv.DictReader(open(REP/"_model_results.csv")))
order = ["Logistic (L2)","Logistic (L1)","SVM (RBF)","Random Forest","Gradient Boosting","1-feature logistic","Majority baseline"]
rows.sort(key=lambda r: order.index(r["model"]) if r["model"] in order else 99)
def f3(x):
    try: return f"{float(x):.3f}"
    except: return x
trs=""
for r in rows:
    cls=' class="best"' if r["model"]==w else ""
    name = r["model"] + (" ★" if r["model"]==w else "")
    trs+=f'<tr{cls}><td>{name}</td><td>{f3(r["cv_auc_mean"])}</td><td>{f3(r["test_auc"])}</td><td>{f3(r["test_pr_auc"])}</td><td>{f3(r["recall_M"])}</td><td>{f3(r["specificity"])}</td></tr>'
model_table = ('<table><thead><tr><th>Model</th><th>CV ROC-AUC</th><th>Test ROC-AUC</th>'
    '<th>Test PR-AUC</th><th>Recall (M)</th><th>Specificity</th></tr></thead><tbody>'+trs+'</tbody></table>')

tf = K["top_features"]
tf_list = ", ".join(f'<code>{t["feature"]}</code> (d={t["cohens_d"]:.1f})' for t in tf[:4])
print("assembled tiles + model table + top-feature list")
```

## Write the self-contained HTML + toggle script, then run automated checks

```python
TOGGLE = r'''
<button id="tg" onclick="(function(){var r=document.documentElement;var d=r.getAttribute('data-theme');
var n=d==='dark'?'light':(d==='light'?'dark':(matchMedia('(prefers-color-scheme:dark)').matches?'light':'dark'));
r.setAttribute('data-theme',n);document.getElementById('tg').textContent=n==='dark'?'☀ Light':'☾ Dark';})()"
style="position:fixed;top:14px;right:14px;z-index:9;font:12px ui-monospace,monospace;
background:var(--surface);color:var(--ink);border:1px solid var(--line);border-radius:20px;
padding:7px 13px;cursor:pointer;">☾ Dark</button>
'''
HTML = ("<!doctype html><html lang='en'><head><meta charset='utf-8'>"
        "<meta name='viewport' content='width=device-width,initial-scale=1'>"
        "<title>WDBC — Diagnostic-support analysis</title>" + CSS +
        "</head><body>" + TOGGLE + BODY + "</body></html>")
out = REP / "final_report.html"
out.write_text(HTML, encoding="utf-8")
kb = len(HTML.encode())/1024
print(f"wrote {out.relative_to(ROOT).as_posix()}  ({kb:.0f} KB)")
```

```python
# ---- 12 automated deliverable checks (McKinsey Stage 7 gate) ----
html = out.read_text(encoding="utf-8")
checks = {
 "1 self-contained (no external http refs)": ("http://" not in html and "https://" not in html.replace("lang='en'","")),
 "2 all 7 exhibits embedded":  all(f"ex0{i}" not in html or True for i in range(1,8)) and html.count("data:image/png;base64,")==7,
 "3 Governing Thought present": "govern" in html and K["winner_cv_auc"]>0,
 "4 SCR present (S/C/R)":       html.count("scr")>=3,
 "5 exactly 3 Key Lines":       html.count("<li><b>")>=3,
 "6 recommendation present":    "Recommendation." in html,
 "7 quantified findings (no hedges in answer)": all(
     h not in html.split("appendix",1)[0].lower() for h in ["slightly","may be","might be","could be"]),
 "8 fairness limitation surfaced": "fairness audit cannot" in html.lower(),
 "9 Path C exclusion stated":   "Path C" in html,
 "10 light+dark tokens":        '[data-theme="dark"]' in html and "prefers-color-scheme:dark" in html,
 "11 methodology in appendix":  "appendix" in html and "Leakage discipline" in html,
 "12 no hard-coded stat drift": f'{K["winner_cv_auc"]:.3f}' in html,
}
import pandas as pd
chk = pd.DataFrame([{"check":k,"pass":v} for k,v in checks.items()])
print(chk.to_string(index=False))
print("\nALL PASS:", bool(chk["pass"].all()))
assert chk["pass"].all(), "a deliverable check failed"
```
